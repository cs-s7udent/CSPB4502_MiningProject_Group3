Create series for each language and then will become columns for further anaylsis

```python
# this method takes a given dataframe and dic and creates the language columns
def extract_lang_per_column(dataframe, all_lang_dic):
    
    # series to hold all languages
    apl = pd.Series()
    ada = pd.Series()
    apex = pd.Series()
    assem = pd.Series()
    bash = pd.Series()
    c = pd.Series()
    c_s = pd.Series()
    c_p = pd.Series()
    c_p1 = pd.Series()
    cobol = pd.Series()
    clo = pd.Series()
    coffee = pd.Series()
    com = pd.Series()
    cry = pd.Series()
    da = pd.Series()
    de = pd.Series()
    el = pd.Series()
    er = pd.Series()
    f_s = pd.Series()
    flow = pd.Series()
    fort = pd.Series()
    gd = pd.Series()
    go = pd.Series()
    gr = pd.Series()
    ht = pd.Series()
    hac = pd.Series()
    has = pd.Series()
    java = pd.Series()
    js = pd.Series()
    ju = pd.Series()
    ko = pd.Series()
    lisp = pd.Series()
    lua = pd.Series()
    mat = pd.Series()
    mp = pd.Series()
    nim = pd.Series()
    objc = pd.Series()
    ocam = pd.Series()
    php = pd.Series()
    perl = pd.Series()
    prolog = pd.Series()
    py = pd.Series()
    r = pd.Series()
    raku = pd.Series()
    ruby = pd.Series()
    rust = pd.Series()
    sas = pd.Series()
    sql = pd.Series()
    scala = pd.Series()
    s_talk = pd.Series()
    solid = pd.Series()
    swift = pd.Series()
    typeS = pd.Series()
    vbnet = pd.Series()
    vba = pd.Series()
    vb = pd.Series()
    weba = pd.Series()
    zig = pd.Series()


    for i in range(len(dataframe)):
        arr = dataframe['tools'][i].strip("[]").replace("'","").split(", ")
        for j in arr:

            if j == "APL":
                apl[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == "Ada":
                ada[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == "Apex":
                apex[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == "Assembly":
                assem[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
        
            # case Bash/Shell, Bash/Shell (all shells), Bash/Shell/PowerShell, PowerShell  => Bash/Shell/PowerShell
            elif j == "Bash/Shell" or j == "Bash/Shell (all shells)"or j == "Bash/Shell/PowerShell" or j =="PowerShell":
                bash[i] = 1
                old = all_lang_dic["Bash/Shell/PowerShell"]
                all_lang_dic["Bash/Shell/PowerShell"] = old + 1
                    
            elif j == "C":
                c[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == "C#":
                c_s[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == "C++":
                c_p[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == "C++11":
                c_p1[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1

            # COBOL, Cobol  => converge into "COBOL"
            elif j == "COBOL" or j == "Cobol":
                cobol[i] = 1
                old = all_lang_dic["COBOL"]
                all_lang_dic["COBOL"] = old + 1

            elif j == "Clojure":
                clo[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == "CoffeeScript":
                coffee[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == "Common Lisp":
                com[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == "Crystal":
                cry[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1

            # Delphi, Delphi/Object Pascal => converge into "Delphi/Object Pascal"
            elif j == "Delphi" or j == "Delphi/Object Pascal":
                de[i] = 1
                old = all_lang_dic["Delphi/Object Pascal"]
                all_lang_dic["Delphi/Object Pascal"] = old + 1

            elif j == "Dart":
                da[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Elixir':
                el[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Erlang':
                er[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'F#':
                f_s[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == "Flow":
                flow[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Fortran':
                fort[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'GDScript':
                gd[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Go':
                go[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Groovy':
                gr[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            
            # HTML,HTML/CSS,CSS => converge into "HTML/CSS"
            elif j == "HTML" or j == "HTML/CSS" or j == "CSS":
                ht[i] = 1
                old = all_lang_dic['HTML/CSS']
                all_lang_dic['HTML/CSS'] = old + 1
            
            elif j == 'Hack':
                hac[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Haskell':
                has[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1     
            if j == 'Java':
                java[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == "JavaScript":
                js[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Julia':
                ju[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Kotlin':
                ko[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            
            # LISP, Lisp => converge into "LISP"
            elif j == 'LISP' or j == 'Lisp':
                lisp[i] = 1
                old = all_lang_dic["LISP"]
                all_lang_dic["LISP"] = old + 1
            
            elif j == "Lua":
                lua[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            
            # MATLAB, Matlab  => converge into "MATLAB"
            elif j == 'MATLAB' or j == 'Matlab':
                mat[i] = 1
                old = all_lang_dic['MATLAB']
                all_lang_dic['MATLAB'] = old + 1        
            
            elif j == 'MicroPython':
                mp[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Nim':
                nim[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Objective-C':
                objc[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1

            # Ocaml,OCaml => converge into "Ocaml"
            elif j == 'Ocaml' or j == 'OCaml':
                ocam[i] = 1
                old = all_lang_dic['Ocaml']
                all_lang_dic['Ocaml'] = old + 1
            
            elif j == "PHP":
                php[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Perl':
                perl[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Prolog':
                prolog[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Python':
                py[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'R':
                r[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1      
            elif j == 'Raku':
                raku[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Ruby':
                ruby[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1     
            elif j == 'Rust':
                rust[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == "SAS":
                sas[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'SQL':
                sql[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1     
            elif j == 'Scala':
                scala[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Smalltalk':
                s_talk[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == 'Solidity':
                solid[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1     
            elif j == 'Swift':
                swift[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            elif j == "TypeScript":
                typeS[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1

            # VB.NET, Visual Basic (.Net)  => converge into "VB.NET"
            elif j == 'VB.NET' or j =='Visual Basic (.Net)':
                vbnet[i] = 1
                old = all_lang_dic['VB.NET']
                all_lang_dic['VB.NET'] = old + 1

            elif j == 'VBA':
                vba[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1
            
            # Visual Basic, Visual Basic 6 => converge into "Visual Basic"      
            elif j == 'Visual Basic' or j == 'Visual Basic 6':
                vb[i] = 1
                old = all_lang_dic['Visual Basic']
                all_lang_dic['Visual Basic'] = old + 1

            elif j == 'WebAssembly':
                weba[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1        
            elif j == 'Zig':
                zig[i] = 1
                old = all_lang_dic[j]
                all_lang_dic[j] = old + 1

           
    dataframe['APL'] = apl
    dataframe['Ada'] = ada
    dataframe['Apex'] = apex
    dataframe['Assembly'] = assem
    dataframe["Bash"] = bash
    dataframe['C'] = c
    dataframe['C#'] = c_s
    dataframe['C++'] = c_p
    dataframe['C++11'] = c_p1
    dataframe['COBOL'] = cobol
    dataframe['Clojure'] = clo
    dataframe['CoffeeScript'] = coffee
    dataframe['Common Lisp'] = com
    dataframe['Crystal'] = cry
    dataframe['Dart'] = da
    dataframe["Delphi/Object_Pascal"] = de
    dataframe['Elixir'] = el
    dataframe['Erlang'] = er
    dataframe['F#'] = f_s
    dataframe['Flow'] = flow
    dataframe['Fortran'] = fort
    dataframe['GDScript'] = gd
    dataframe['Go'] = go
    dataframe['Groovy'] = gr
    dataframe['HTML/CSS'] = ht
    dataframe['Hack'] = hac
    dataframe['Haskell'] = has
    dataframe['Java'] = java
    dataframe['JavaScript'] = js
    dataframe['Julia'] = ju
    dataframe['Kotlin'] = ko
    dataframe['LISP'] = lisp
    dataframe['Lua'] = lua
    dataframe['MATLAB'] = mat
    dataframe['MicroPython'] = mp
    dataframe['Nim'] = nim
    dataframe['Objective-C'] = objc
    dataframe['Ocaml'] = ocam
    dataframe['PHP'] = php
    dataframe['Perl'] = perl
    dataframe['Prolog'] = prolog
    dataframe['Python'] = py
    dataframe['R'] = r
    dataframe['Raku'] = raku
    dataframe['Ruby'] = ruby
    dataframe['Rust'] = rust
    dataframe['SAS'] = sas
    dataframe['SQL'] = sql
    dataframe['Scala'] = scala
    dataframe['Smalltalk'] = s_talk
    dataframe['Solidity'] = solid
    dataframe['Swift'] = swift
    dataframe['TypeScript'] = typeS
    dataframe['VB.NET'] = vbnet
    dataframe['VBA'] = vba
    dataframe['Visual_Basic'] = vb
    dataframe['WebAssembly'] = weba
    dataframe['Zig'] = zig


    
```

```python
# this method takes a given dataframe column and extracts the list values into the given dictionary
def get_all_lang(df_column, dic):
    

    for entry in df_column:
        # print(entry)
        if type(entry) != str:
            entry = entry.asType(str)
        entry = entry.strip("[]").replace("'","").split(", ")
        for i in entry:
            i = i.lower()
            if i not in dic:
                dic[i] = 1
            else:
                old = dic[i]
                dic[i] = old + 1
    
    return dic
```
