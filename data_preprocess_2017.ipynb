```python
# this method loops thru every value of the given pandas series and calculates
# the total number of nan values and the fraction of 
# the total number of nan values / the total values in series 
def count_nan(series):
    count = 0
    for i in series:
        if i is np.nan:
            count += 1
    return count, count/len(series)
```

```python
# this method transforms a string that contains different values separated by
# a semicolon and a space(; ) of a given attribute in a given dataframe into a list and 
# returns in the pandas series format and a nparray that contains the indexes of rows that
# has nan 
def merge_str_list(df, attr_name):
    drop_lst = np.array([])
    result = pd.Series()
    for i in range(df.shape[0]):
        data = df[attr_name][i]
        # if string value, extract all values into a list
        if type(data) is str:
            result[i] = data.split('; ')
        # if not string value, then add the index into drop_lst
        else:
            drop_lst = np.append(drop_lst, i)
            result[i] = data
    
    return result, drop_lst
```

```python
# calculate how many respondents wrote nan in developer type and wrote non-nan value in nondeveloper type

dev_nan_lst = np.array([])
data = df['DeveloperType']
for i in range(len(data)):
    if data[i] is np.nan:
        dev_nan_lst = np.append(dev_nan_lst, i)

non_dev_lst = np.array([])
data = df['NonDeveloperType']
for i in range(len(data)):
    if data[i] is not np.nan:
        non_dev_lst = np.append(non_dev_lst, i)

print("The total number of nan values in Developertype attribute is " + str(len(dev_nan_lst)))
print("The total number of non developer respondents is " + str(len(non_dev_lst)))
common_lst = np.intersect1d(dev_nan_lst, non_dev_lst)
print("There are total number of respondents that have nan values in Developertype attribute but has selected a non-nan value"+
      " in non developer is " + str(len(common_lst)))
print(", and therefore we are going to create a new attribute developtype_t and fill in the nan values")
```

For df['DeveloperType'], we need to merge df['WebDeveloperType'] into row value = Web developer (front-end, back-end, or full stack) in df['DeveloperType'], merge df['MobileDeveloperType'] into row value = Mobile developer and fill in df['NonDeveloperType'] into row value = nan. These new values will be in a new attribute call developertype_t

```python
developertype_t = pd.Series()
data_dev = df['DeveloperType']
data_web = df['WebDeveloperType']
data_mobile = df['MobileDeveloperType']
data_non_dev = df['NonDeveloperType']
for i in range(len(data_dev)): 
    curr_data = data_dev[i]
    lst = list()
    # if string, separate all values and add them in a list
    if type(curr_data) == str:
        a = curr_data.split("; ")
        for j in a:
            # web developer only single choice

            print(j)
            if j == 'Web developer':
                d = j
                if type(data_web[i]) == str:
                    d += "_" + data_web[i]
                lst.append(d)    

            # mobile developer is multiple choice
            elif j == 'Mobile developer':
                d = j
                if type(data_mobile[i]) == str:
                    b = data_mobile[i].split("; ")
                    for k in b:
                        d += "_" + k
                
                lst.append(d)

            else:
                lst.append(j)
        developertype_t[i] = lst
    # if not string, see if non developertype has data 
    else:
        if type(data_non_dev[i]) == str:
            c = data_non_dev[i].split("; ")
            for l in c:
                if l == "C-suite executive (CEO, COO, CTO, CFO, etc.)":
                    l = "C-suite executive (CEO/COO/CTO/CFO/etc.) "
                lst.append(l)
            developertype_t[i] = lst
            
        else:
            developertype_t[i] = curr_data
        
        
# create a new attribute
df['developertype_t'] = developertype_t
```

Need to change the years program into the following range

Old value                                                  => New value

Less than a year, 1 to 2 years                                         => 0 - 2 years

2 to 3 years, 3 to 4 years, 4 to 5 years                   => 2 - 5 years

5 to 6 years, 6 to 7 years, 7 to 8 years, 9 to 10 years    => 6 - 10 years

10 to 11 years, 11 to 12 years, 12 to 13 years, 
13 to 14 years, 15 to 16 years, 16 to 17 years,
17 to 18 years, 18 to 19 years, 19 to 20 years,
20 or more years                                           => 11+ years


```python
# change bin size and create a new attribute 'code_yrs'
data = df['YearsProgram']
code_yrs = pd.Series()
for i in range(len(data)):
    if data[i] == "Less than a year" or data[i] == "1 to 2 years":
        code_yrs[i] = "0 - 2 years"
    elif data[i] == "2 to 3 years" or data[i] == "3 to 4 years" or data[i] == "4 to 5 years":
        code_yrs[i] = "2 - 5 years"
    elif data[i] ==  "5 to 6 years" or data[i] == "6 to 7 years" or data[i] == "7 to 8 years" or data[i] == "8 to 9 years" or data[i] == "9 to 10 years":
        code_yrs[i] = "6 - 10 years"
    # for all other values that are not nan, put them in 11+ years
    else:
        if data[i] is not np.nan:
            code_yrs[i] = "11+ years"

# create a new attribute
df['code_yrs'] = code_yrs
```

Transform float values of Salary into ranges as below:

Less than $20,000         
$20,000 - $40,000          
$40,000 - $60,000          
$60,000 - $80,000          
$80,000 - $100,000         
$100,000 - $120,000        
Rather not say             
Unemployed                 
More than $160,000         
$140,000 - $160,000        
$120,000 - $140,000         
Other (please specify)      

There will be no values for Rather not say, Unemployed and Other

```python
# this transform the values of salary into the same bin size as previous surveys and create a new attribute salary_merge
salary_merge = pd.Series()
data = df['Salary']
for i in range(len(data)):
    if data[i] < 20000:
        salary_merge[i] = "Less than $20,000"
    elif data[i] >= 20000 and data[i] < 40000:
        salary_merge[i] = "$20,000 - $40,000"
    elif data[i] >= 40000 and data[i] < 60000:
        salary_merge[i] = "$40,000 - $60,000"
    elif data[i] >= 60000 and data[i] < 80000:
        salary_merge[i] = "$60,000 - $80,000"
    elif data[i] >= 80000 and data[i] < 100000:
        salary_merge[i] = "$80,000 - $100,000"
    elif data[i] >= 100000 and data[i] < 120000:
        salary_merge[i] = "$100,000 - $120,000"
    elif data[i] >= 120000 and data[i] < 140000:
        salary_merge[i] = "$120,000 - $140,000"
    elif data[i] >= 140000 and data[i] < 160000:
        salary_merge[i] = "$140,000 - $160,000"
    elif data[i] >= 160000:
        salary_merge[i] = "More than $160,000"
    # if nan, the transformed data is still nan
    elif np.isnan(data[i]):
        salary_merge[i] = np.nan

df['salary_merge'] = salary_merge
```

Need to combine values in df['HaveWorkedDatabase'], df['HaveWorkedFramework'], df['HaveWorkedLanguage'] and df['HaveWorkedPlatform'] and merge them into a new attribute called tools.
If nan values are in all four attributes, then drop that row.

```python
# merge all four attributes into one new attribute called 'tools' 
# and extract the row indexes that have nan values in all four attributes
data_db = df['HaveWorkedDatabase']
data_fw = df['HaveWorkedFramework']
data_lang = df['HaveWorkedLanguage']
data_plat = df['HaveWorkedPlatform']
tools = pd.Series()
drop_lst = np.array([])
for i in range(len(df)):
    a = data_db[i]
    b = data_fw[i]
    c = data_lang[i]
    d = data_plat[i]
    a_lst = list()
    b_lst = list()
    c_lst = list()
    d_lst = list()
    if type(a) == str:
        a_lst = a.split("; ")
    if type(b) == str:
        b_lst = b.split("; ")
    if type(c) == str:
        c_lst = c.split("; ")
    if type(d) == str:
        d_lst = d.split("; ")
    # if four attributes are all nan values, then add the row index into drop list    
    if len(a_lst) == 0 and len(b_lst) == 0 and len(c_lst) == 0 and len(d_lst) == 0 :
        tools[i] = np.nan
        drop_lst = np.append(drop_lst, i)
    else:
        tools[i] = a_lst + b_lst + c_lst + d_lst

# create new attribute tools
df['tools'] = tools
```

```python
# drop the rows that have all nan values and create a new datatframe df2
print("Rows at these indexes have all nan value and need to be dropped \n "+str(drop_lst))
print("The total number of nan values in all four attributes \n('HaveWorkedDatabase','HaveWorkedFramework','HaveWorkedLanguage' and 'HaveWorkedPlatform')"
      +" \n is "+str(len(drop_lst))+" which the proportion is "+str(len(drop_lst)/len(df)))
df2 = df.drop(index = drop_lst)
```
