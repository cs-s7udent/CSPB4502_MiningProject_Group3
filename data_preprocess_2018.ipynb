```python
# this method loops thru every value of the given pandas series and calculates
# the total number of nan values and the fraction of 
# the total number of nan values / the total values in series 
def count_nan(series):
    count = 0
    for i in series:
        if i is np.nan:
            count += 1
    return count, count/len(series)
```

```python
# this method transforms a string that contains different values separated by
# a semicolon and a space(;) of a given attribute in a given dataframe into a list and 
# returns in the pandas series format and a nparray that contains the indexes of rows that
# has nan 
def merge_str_list(df, attr_name):
    drop_lst = np.array([])
    result = pd.Series()
    for i in range(df.shape[0]):
        data = df[attr_name][i]
        # if string value, extract all values into a list
        if type(data) is str:
            result[i] = data.split(';')
        # if not string value, then add the index into drop_lst
        else:
            drop_lst = np.append(drop_lst, i)
            result[i] = data
    
    return result, drop_lst
```

Need to change the bin size of df['YearsCoding'] into the following:

old value                      => new value

0-2 years                      => 0 - 2 years

3-5 years                      => 2 - 5 years

6-8 years, 9-11 years          => 6 - 10 years

12-14 years, 15-17 years, 18-20 years, 21-23 years, 24-26 years, 27-29 years, 30 or more years   => 11+ years

```python
# change bin size and create a new attribute 'code_yrs'
data = df['YearsCoding']
code_yrs = pd.Series()
for i in range(len(data)):
    if data[i] == "0-2 years":
        code_yrs[i] = "0 - 2 years"
    elif data[i] == "3-5 years":
        code_yrs[i] = "2 - 5 years"
    elif data[i] == "6-8 years" or data[i] == "9-11 years":
        code_yrs[i] = "6 - 10 years"
    # for all other values that are not nan, put them in 11+ years
    else:
        if data[i] is not np.nan:
            code_yrs[i] = "11+ years"

# create a new attribute
df['code_yrs'] = code_yrs
```

Merge df['FormalEducation'], df['UndergradMajor'] and df['EducationTypes'] into a new attribute 'edu_t'

```python
# Merge df['FormalEducation'], df['UndergradMajor'] and df['EducationTypes'] into a new attribute 'edu_t'
data_formal = df['FormalEducation']
data_maj = df['UndergradMajor']
data_learn = df['EducationTypes']
edu_t = pd.Series()

for i in range(len(df)):
    a = data_formal[i]
    b = data_maj[i]
    c = data_learn[i]
    lst = list()

    
    if a is not np.nan:
        #  a and b are not nan
        if b is not np.nan:
            lst.append(a)
            lst.append(b)
            if type(c) is str:
                c_lst = c.split(";")
                for j in c_lst:
                    lst.append(j)
            edu_t[i] = lst
        # a is not nan and b is nan
        else:
            lst.append(a)
            if type(c) is str:
                c_lst = c.split(";")
                for j in c_lst:
                    lst.append(j)
            edu_t[i] = lst
    # a is nan            
    else:
        if type(c) is str:
            c_lst = c.split(";")
            for j in c_lst:
                lst.append(j)
            edu_t[i] = lst    
        else:
            edu_t[i] = a
    

df['edu_t'] = edu_t
```

Transform float values of Converted Salary into ranges as below:

Less than $20,000         
$20,000 - $40,000          
$40,000 - $60,000          
$60,000 - $80,000          
$80,000 - $100,000         
$100,000 - $120,000        
Rather not say             
Unemployed                 
More than $160,000         
$140,000 - $160,000        
$120,000 - $140,000         
Other (please specify)      

There will be no values for Rather not say, Unemployed and Other

```python
# this transform the values of salary into the same bin size as previous surveys and create a new attribute salary_merge
salary_merge = pd.Series()
data = df['ConvertedSalary']
for i in range(len(data)):
    if data[i] < 20000:
        salary_merge[i] = "Less than $20,000"
    elif data[i] >= 20000 and data[i] < 40000:
        salary_merge[i] = "$20,000 - $40,000"
    elif data[i] >= 40000 and data[i] < 60000:
        salary_merge[i] = "$40,000 - $60,000"
    elif data[i] >= 60000 and data[i] < 80000:
        salary_merge[i] = "$60,000 - $80,000"
    elif data[i] >= 80000 and data[i] < 100000:
        salary_merge[i] = "$80,000 - $100,000"
    elif data[i] >= 100000 and data[i] < 120000:
        salary_merge[i] = "$100,000 - $120,000"
    elif data[i] >= 120000 and data[i] < 140000:
        salary_merge[i] = "$120,000 - $140,000"
    elif data[i] >= 140000 and data[i] < 160000:
        salary_merge[i] = "$140,000 - $160,000"
    elif data[i] >= 160000:
        salary_merge[i] = "More than $160,000"
    # if nan, the transformed data is still nan
    elif np.isnan(data[i]):
        salary_merge[i] = np.nan

df['salary_merge'] = salary_merge
```

Need to combine values in df['LanguageWorkedWith'], df['DatabaseWorkedWith'] and df['PlatformWorkedWith'] and df['FrameworkWorkedWith'] and merge them into a new attribute called tools.
If nan values are in all four attributes, then drop that row.


```python
# merge all four attributes into one new attribute called 'tools' and extract the row indexes that have nan values in all three attributes
data_db = df['DatabaseWorkedWith']
data_fw = df['FrameworkWorkedWith']
data_lang = df['LanguageWorkedWith']
data_plat = df['PlatformWorkedWith']
tools = pd.Series()
drop_lst = np.array([])
for i in range(len(df)):
    a = data_db[i]
    b = data_fw[i]
    c = data_lang[i]
    d = data_plat[i]
    a_lst = list()
    b_lst = list()
    c_lst = list()
    d_lst = list()
    if type(a) == str:
        a_lst = a.split(";")
    if type(b) == str:
        b_lst = b.split(";")
    if type(c) == str:
        c_lst = c.split(";")
    if type(d) == str:
        d_lst = d.split(";")
    # if four attributes are all nan values, then add the row index into drop list    
    if len(a_lst) == 0 and len(b_lst) == 0 and len(c_lst) == 0 and len(d_lst) == 0 :
        tools[i] = np.nan
        drop_lst = np.append(drop_lst, i)
    else:
        tools[i] = a_lst + b_lst + c_lst + d_lst

# create new attribute tools
df['tools'] = tools
```

```python
# drop the rows that have all nan values and create a new datatframe df2
print("Rows at these indexes have all nan value and need to be dropped \n "+str(drop_lst))
print("The total number of nan values in all four attributes \n('DatabaseWorkedWith','FrameworkWorkedWith','LanguageWorkedWith' and 'PlatformWorkedWith')"
      +" \n is "+str(len(drop_lst))+" which the proportion is "+str(len(drop_lst)/len(df)))
df2 = df.drop(index = drop_lst)
```
