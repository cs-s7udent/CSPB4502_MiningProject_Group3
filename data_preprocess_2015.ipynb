```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
# this method loops thru every value of the given pandas series and calculates
# the total number of nan values and the fraction of 
# the total number of nan values / the total values in series 
def count_nan(series):
    count = 0
    for i in series:
        if i is np.nan:
            count += 1
    return count, count/len(series)
```

```python
# this method extracts values of the given attributes indexes, saves into a separate list for each row and 
# returns a series of lists for all rows and a nparray of row indexs that have all nan values
def merge_attribute(df, attr_index_start, attr_index_end): 
    tools = pd.Series()
    nan_arr = np.array([])
    for i in range (df.shape[0]):
        items = list()
        for c in range(attr_index_start, attr_index_end+1):
            curr = df[df.columns[c]][i]
            if type(curr) == str:
                items.append(curr)
        # if a certain row has nan values for all the given attributes, add the row index into nan_arr
        # and set the value to np.nan in tools
        if len(items) == 0:
            nan_arr = np.append(nan_arr, i)
            tools[i] = np.nan
        else:
            tools[i] = items
          
    return tools, nan_arr
```

```python
# read csv
df = pd.read_csv("2015 Stack Overflow Developer Survey Responses.csv", skiprows=1)
df
```

```python
# print out all attributes
total_attributes = df.columns
for i in total_attributes:
    print(i)
```

```python
# merge all Training & Education attributes into a new attribute called 'education'
# education related attributes start at index 94 to index 104 for 2015 survey
df['education'], edu_nan_lst = merge_attribute(df, 94, 104)
```

```python
# merge all Current Lang & Tech attributes and Source Control into a new attribute called 'tools'
# attributes of current language and tools start at index 8 to index 50 for 2015 survey
# attributes of source control start at index 191 to index 201 for 2015 survey

tools = pd.Series()
sourceControl, drop_a = merge_attribute(df, 191, 201)
langTech, drop_b = merge_attribute(df, 8, 50)
for i in range(len(sourceControl)):
    lst = list()
    a_list = langTech[i]
    b_list = sourceControl[i]

    if a_list is not np.nan:
        for j in a_list:
            # replace "Cloud (AWS, GAE, Azure, etc.)" to "Cloud (AWS/GAE/Azure/etc.)"
            if j == "Cloud (AWS, GAE, Azure, etc.)":
                lst.append("Cloud (AWS/GAE/Azure/etc.)")
            else:
                lst.append(j)
    if b_list is not np.nan:
        for k in b_list:
            lst.append(k)

    # if both a and b are nan
    if len(lst) == 0:
        tools[i] = np.nan
    # otherwise, add lst to tools
    else:
        tools[i] = lst

# drop list is the intersection of drop_a and drop_b
drop_lst = np.intersect1d(drop_a, drop_b)

# create new attribute
df['tools'] = tools
```

```python
# change the bin size of df2['Years IT / Programming Experience'] to match other surveys (2018)
# create a new attribute years_t
yrs = pd.Series()
data = df['Years IT / Programming Experience']
for i in range(len(data)):
    if data[i] == "Less than 1 year" or data[i] == "1 - 2 years":
        yrs[i] = "0 - 2 years"
    else:
        yrs[i] = data[i]

df['years_t'] = yrs
```

```python
# drop the rows that have all nan values for current language and tools and set the new dataframe as df2
print("Rows at these indexes have all nan value and need to be dropped "+str(drop_lst))
print("There are total " + str(len(drop_lst)) + " rows that need to be dropped")
df2 = df.drop(index=drop_lst)
```
