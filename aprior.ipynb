```python
# This method takes a support level and a dataframe column and prints out the apriori itemsets
# Please note that this method only supports to k=3 datasets
def apriori(support_level, df_column):
    go = True
    k = 1
    min_support = len(df_column) * support_level
    print("total count is " + str(len(df_column)))
    print("support level is " + str(support_level) + " and min support is " + str(min_support))
    
    sorted_lst = []
    dic1 = dict()
    dic2 = dict()
    dic3 = dict()
    dic4 = dict()
    dic5 = dict()
    dic6 = dict()

    
    # extract entries from df_column and create the dic for k=1 itemset
    for entry in df_column:
        # print(type(entry))
        # print(entry)
        if type(entry) != str:
            entry = entry.asType(str)
        entry = entry.strip("[]").replace("'","").split(", ")
        entry.sort()
        
        # append the sorted entry into a new list called sorted_lst
        sorted_lst.append(entry)
        
        for key in entry:
            if key not in dic1:
                dic1[key] = 1
            else:
                old = dic1[key]
                dic1[key] = old + 1

    print("Dic is " + str(dic1))

    ok_lst = []        
    for i in dic1:
        if dic1[i] >= min_support:
            ok_lst.append(i)

    # if no min support is met for k=1 itemset, then stop   
    if (len(ok_lst) == 0):
        go = False
            
    ok_lst.sort()

    # print out k=1 itemlist
    # print("k = " + str(k) + " itemset list: ")
    # for i in ok_lst:
    #     print("['" + str(i) + "']")

    dic = dict()
    # otherwise continue to k=2+
    while (go):
        # increment k
        k = k + 1       
        
        # get the right dict
        if k == 2:
            dic = dic2
        elif k == 3:
            dic = dic3
        elif k == 4:
            dic = dic4
        elif k == 5:
            dic = dic5
        elif k == 6:
            dic = dic6
        
        x = []

        prev_ok_lst = ok_lst

        # print("start of while loop: k = " +str(k))
        for i in ok_lst:
            lst = i.split("_")
            for j in lst:
                # print(j)
                x.append(j)
        y = set(x)
        y = list(y)
        y.sort()
        # print("y: "+str(y))
        
        ok_lst = y

        key = ""
        first_key_w = ok_lst[0]

        # get key
        # when the k is the same as the number of itemset
        if k == len(ok_lst):
            for i in range(len(ok_lst)):
                if i == len(ok_lst) - 1:
                    key += ok_lst[i]
                else:
                    key += ok_lst[i]
                    key += "_"
            
            dic[key] = 0
       
        # get key
        # when itemset - k = 1 
        # elif len(ok_lst) - k == 1:
        #     for i in range(len(ok_lst)-1, -1, -1):
        #         key_count = 1
        #         key = ""
        #         for j in range(len(ok_lst)):
        #             if i != j:
        #                 if key_count == k:
        #                     key += ok_lst[j]
        #                 else:
        #                     key += ok_lst[j]
        #                     key += "_"
        #                     key_count += 1
                    
        #         dic[key] = 0                

        # get key
        # when k = 2
        elif k == 2:
            for i in range(len(ok_lst)-1):
                for j in range(i+1, len(ok_lst)):
                    key = ok_lst[i] + "_" + ok_lst[j]
                    dic[key] = 0

        # get key
        # when k = 3
        elif k == 3:
            for i in range(len(ok_lst)-2):
                for j in range(i+1, len(ok_lst)-1):
                    for l in range(j+1, len(ok_lst)):
                        key = ok_lst[i] + "_" + ok_lst[j] + "_" + ok_lst[l]
                        dic[key] = 0

        # get key
        # when k = 4
        elif k == 4:
            for i in range(len(ok_lst)-3):
                for j in range(i+1, len(ok_lst)-2):
                    for l in range(j+1, len(ok_lst)-1):
                        for m in range(l+1, len(ok_lst)):
                            key = ok_lst[i] + "_" + ok_lst[j] + "_" + ok_lst[l] + "_" + ok_lst[m]
                            dic[key] = 0


        # get key
        # when k = 5
        elif k == 5:
            for i in range(len(ok_lst)-4):
                for j in range(i+1, len(ok_lst)-3):
                    for l in range(j+1, len(ok_lst)-2):
                        for m in range(l+1, len(ok_lst)-1):
                            for n in range(m+1, len(ok_lst)):
                                key = ok_lst[i] + "_" + ok_lst[j] + "_" + ok_lst[l] + "_" + ok_lst[m] + "_" + ok_lst[n]
                                dic[key] = 0
            

        # get key
        # when k = 6
        elif k == 6:
            for i in range(len(ok_lst)-5):
                for j in range(i+1, len(ok_lst)-4):
                    for l in range(j+1, len(ok_lst)-3):
                        for m in range(l+1, len(ok_lst)-2):
                            for n in range(m+1, len(ok_lst)-1):
                                for o in range(n+1, len(ok_lst)):
                                    key = ok_lst[i] + "_" + ok_lst[j] + "_" + ok_lst[l] + "_" + ok_lst[m] + "_" + ok_lst[n] + "_" + ok_lst[o]
                                    dic[key] = 0


        else:
            print("ONLY SUPPORT TO K=6 ITEMSETS")
            go = False


        # loop thru dataset and fill in dic
        # when k = 2
        if k==2:       
            for entries in sorted_lst:
                # print(entries)
                for i in range(len(entries)-1):
                    for j in range(i+1, len(entries)):
                
                        key = entries[i]+"_"+entries[j]

                        if key in dic:

                            old = dic[key]
                            dic[key] = old + 1
        
        # loop thru dataset and fill in dic
        # when k = 3
        elif k == 3:
            for entries in sorted_lst:
                start_i = 0
                keep_going = False 
                # get the index of the first key word (to speed up process)
                for i in range(len(entries)):
                    # print("first key w " + str(first_key_w))
                    # print(entries[i])
                    
                    # continue if first key word can be found in entry
                    if (entries[i] == first_key_w):
                        start_i = i
                        keep_going = True
                        break
                    # otherwise akip
                    elif (entries[i] > first_key_w):
                        break
                                   
                if keep_going:
                    for i in range(start_i, len(entries)-2):
                        for j in range(i+1, len(entries)-1):
                            for l in range(j+1, len(entries)):
                                key = entries[i] + "_" +  entries[j] + "_" + entries[l]
                                # print(key)
                                if key in dic:
                                    old = dic[key]
                                    dic[key] = old + 1
            

        # loop thru dataset and fill in dic
        # when k = 4
        elif k == 4:
            for entries in sorted_lst:
                start_i = 0
                keep_going = False 
                # get the index of the first key word (to speed up process)
                for i in range(len(entries)):
                    # print("first key w " + str(first_key_w))
                    # print(entries[i])
                    
                    # continue if first key word can be found in entry
                    if (entries[i] == first_key_w):
                        start_i = i
                        keep_going = True
                        break
                    # otherwise akip
                    elif (entries[i] > first_key_w):
                        break
                                   
                if keep_going:
                    # for key in dic:
                    #     lst = key.split("_")
                    #     if len(np.intersect1d(entries, lst)) == len(lst):
                    #         old = dic[key]
                    #         dic[key] = old + 1



                    for i in range(start_i, len(entries)-3):
                        for j in range(i+1, len(entries)-2):
                            for l in range(j+1, len(entries)-1):
                                for m in range(l+1, len(entries)):
                                    key = entries[i] + "_" +  entries[j] + "_" + entries[l] + "_" + entries[m]
                                    # print(key)
                                    if key in dic:
                                        old = dic[key]
                                        dic[key] = old + 1
        
        
        # loop thru dataset and fill in dic
        # when k = 5
        elif k == 5:

            for entries in sorted_lst:
                start_i = 0
                keep_going = False 
                # get the index of the first key word (to speed up process)
                for i in range(len(entries)):
                    # print("first key w " + str(first_key_w))
                    # print(entries[i])
                    
                    # continue if first key word can be found in entry
                    if (entries[i] == first_key_w):
                        start_i = i
                        keep_going = True
                        break
                    # otherwise akip
                    elif (entries[i] > first_key_w):
                        break
                                   
                if keep_going:
                    for key in dic:
                        lst = key.split("_")
                        if len(np.intersect1d(entries, lst)) == len(lst):
                            old = dic[key]
                            dic[key] = old + 1


                    # for i in range(start_i, len(entries)-4):
                    #     for j in range(i+1, len(entries)-3):
                    #         for l in range(j+1, len(entries)-2):
                    #             for m in range(l+1, len(entries)-1):
                    #                 for n in range(m+1, len(entries)):
                    #                     key = entries[i] + "_" +  entries[j] + "_" + entries[l] + "_" + entries[m] + "_" + entries[n]
                    #                     # print(key)
                    #                     if key in dic:
                    #                         old = dic[key]
                    #                         dic[key] = old + 1

        # loop thru dataset and fill in dic
        # when k = 6
        elif k == 6:
            for entries in sorted_lst:
                start_i = 0
                keep_going = False 
                # get the index of the first key word (to speed up process)
                for i in range(len(entries)):
                    # print("first key w " + str(first_key_w))
                    # print(entries[i])
                    
                    # continue if first key word can be found in entry
                    if (entries[i] == first_key_w):
                        start_i = i
                        keep_going = True
                        break
                    # otherwise skip
                    elif (entries[i] > first_key_w):
                        break

                if keep_going:

                    for key in dic:
                        lst = key.split("_")
                        if len(np.intersect1d(entries, lst)) == len(lst):
                            old = dic[key]
                            dic[key] = old + 1

                    # for i in range(start_i, len(entries)-5):
                    #     for j in range(i+1, len(entries)-4):
                    #         for l in range(j+1, len(entries)-3):
                    #             for m in range(l+1, len(entries)-2):
                    #                 for n in range(m+1, len(entries)-1):
                    #                     for o in range(n+1, len(entries)):
                    #                         key = entries[i] + "_" +  entries[j] + "_" + entries[l] + "_" + entries[m] + "_" + entries[n] + "_" + entries[o]
                    #                         # print(key)
                    #                         if key in dic:
                    #                             old = dic[key]
                    #                             dic[key] = old + 1

        else:
            print("ONLY SUPPORT TO K=6 ITEMSETS")
            go = False
        
        print("Dic is " + str(dic))
        
        ok_lst = []

        for i in dic:
            if dic[i] >= min_support:
                ok_lst.append(i)
                prev_ok_lst.append(i)
            

        # if the next itemlist is empty (meaning no items have passed the min support count), stop loop, print out frequent itemlist and calculate confidence level
        if (len(ok_lst) <= 1):
            go = False

            # print out itemlist
            print("k = " + str(k-1) + " itemset list: " + str(prev_ok_lst))
            for i in prev_ok_lst:
                a = i.split("_")
                print(a)

            # check confidence level
                # k = 2
                if (len(a) == 2):
                    # v1 => v2
                    # v2 => v1
                    v1 = dic1[a[0]]
                    v2 = dic1[a[1]]
                    v1v2 = dic2[i]
                    print(str(a[0]) + " => " + str(a[1]) + ", support = " + str(v1v2/len(df_column)))
                    print(str(a[0]) + " => " + str(a[1]) + ", confidence = " + str(v1v2/v1))
                    print(str(a[1]) + " => " + str(a[0]) + ", support = " + str(v1v2/len(df_column)))
                    print(str(a[1]) + " => " + str(a[0]) + ", confidence = " + str(v1v2/v2))


                # k = 3
                elif (len(a) == 3):
                    # (v1, v2, v3)
                    # (v1, v2) => v3
                    # (v1, v3) => v2
                    # (v2, v3) => v1
                    # v1 => (v2, v3)
                    # v2 => (v1, v3)
                    # v3 => (v1, v2)
        
                    v1 = dic1[a[0]]
                    v2 = dic1[a[1]]
                    v3 = dic1[a[2]]

                    v1v2 = dic2[a[0]+"_"+a[1]]
                    v2v3 = dic2[a[1]+"_"+a[2]]
                    v1v3 = dic2[a[0]+"_"+a[2]]

                    v1v2v3 = dic3[i]

                    print(str(a[0])+","+str(a[1])+" => "+str(a[2])+", support = "+str(v1v2v3/len(df_column)))
                    print(str(a[0])+","+str(a[1])+" => "+str(a[2])+", confidence = "+str(v1v2v3/v1v2))
                    print(str(a[0])+","+str(a[2])+" => "+str(a[1])+", support = "+str(v1v2v3/len(df_column)))
                    print(str(a[0])+","+str(a[2])+" => "+str(a[1])+", confidence = "+str(v1v2v3/v1v3))
                    print(str(a[1])+","+str(a[2])+" => "+str(a[0])+", support = "+str(v1v2v3/len(df_column)))
                    print(str(a[1])+","+str(a[2])+" => "+str(a[0])+", confidence = "+str(v1v2v3/v2v3))
                    print(str(a[0])+" => "+str(a[1])+","+str(a[2])+", support = "+str(v1v2v3/len(df_column)))
                    print(str(a[0])+" => "+str(a[1])+","+str(a[2])+", confidence = "+str(v1v2v3/v1))
                    print(str(a[1])+" => "+str(a[0])+","+str(a[2])+", support = "+str(v1v2v3/len(df_column)))
                    print(str(a[1])+" => "+str(a[0])+","+str(a[2])+", confidence = "+str(v1v2v3/v2))
                    print(str(a[2])+" => "+str(a[0])+","+str(a[1])+", support = "+str(v1v2v3/len(df_column)))
                    print(str(a[2])+" => "+str(a[0])+","+str(a[1])+", confidence = "+str(v1v2v3/v3))


                # elif (len(a) == 4):
                    
                    # (v1, v2, v3, v4)
                    # (v1, v2) => v3
                    # (v1, v3) => v2
                    # (v2, v3) => v1
                    # v1 => (v2, v3)
                    # v2 => (v1, v3)
                    # v3 => (v1, v2)


    
```
